```python
from helper_ML_pairtools import GetPairEnergies, GenerateFeatures,pair_energy, CompareLocalization
import os
import time
from helper_CC_ML_spacial import *
from helper_ML_tools import SplitFolder


import psi4
```

```python
# User defined example name
Calc_name='example_name'

# Path to the xyz coordinates with subfiles for train and test data
Folder='xyz_str/'

# Run using Pipek-Mezey localization and the STO-3G basis set considering only 10 of the excitations for each pair-energy
ML=pair_energy(Folder+'train/',test_folder=Folder+'test/', loc_occ='PM', loc_vir='PM', Triples=False, basis='STO-3G', values=10)# cutoff=1e-5,)

```

```python
# Run the machine learning 
ML.test_folder=Folder+'test/'

ML.calculate_train()

ML.train_model()

a,b,c,d=ML.test()
```

```python
# Post calculation data analysis
df_temp=[pd.DataFrame(a), pd.DataFrame(b), pd.DataFrame(c), pd.DataFrame(d)]
df=pd.concat(df_temp,axis=1)

df.columns=['Names', 'Absolute Error', 'True Energy', 'Predicted Energy']

df.to_csv(Calc_name+'.csv')

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['font.family'] = "sans-serif"
matplotlib.rcParams['font.sans-serif'] = "Times New Roman"
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"] + plt.rcParams["font.serif"]
#plt=plt.figsize(12,12)
ax=sns.jointplot(x=c,y=(c-d), color="#4CB391",s=2)
#plt.plot(,)
x0, x1 = ax.ax_joint.get_xlim()
ax.ax_joint.plot([x0,x1], [1/627.509,1/627.509], ':k')  
ax.ax_joint.plot([x0,x1], [-1/627.509,-1/627.509], ':k')  
#axes=fig.axes
#plt.set(ylim=(-.005,.005))
ax.fig.axes[0].set_ylim(-0.003,0.003)
#axes[0,0].set_ylim(0,1)
ax.fig.axes[0].set_xlabel('True Correlation Energy ($E_\mathdefault{h}$)', fontsize=14)
ax.fig.axes[0].set_ylabel('Predicted Energy Error ($E_\mathdefault{h}$)', fontsize=14)
#plt.ylim(0,3)
plt.tight_layout()
plt.savefig(Calc_name+'.png',dpi=300,bbox_inches='tight')

```
